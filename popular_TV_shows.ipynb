```python
import pandas as pd
import requests
```

```python
df = pd.DataFrame()
df
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
for i in range(1, 7535):
    response = requests.get('https://api.themoviedb.org/3/tv/popular?api_key=5efd3acfdcd0bd63456fb3480fdc81a1')
    temp = pd.DataFrame(response.json()['results'])[['first_air_date', 'genre_ids', 'name', 'origin_country', 'original_language', 'overview', 'popularity', 'vote_average', 'vote_count']]
    df = df.append(temp) ##ignore_index=True should be used
```
